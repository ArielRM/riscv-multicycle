```python
import numpy as np
import math
import matplotlib.pyplot as plt
import sys

# Generate n-values (sine_resolution) of a sine wave according to the defined bitsize
# Then print these values to the file sine_values.txt

SINE_RESOLUTION = 1000
SINE_VALUES_BITSIZE = 16
t = 0
sine_values = []

for i in range (0,SINE_RESOLUTION):
    t = ((i*360/(SINE_RESOLUTION))*(math.pi/180))
    sine_values.append(math.sin(t))

sine = np.array(sine_values)
print("Number of points in sine wave: ", np.size(sine))

j = np.linspace(0, SINE_RESOLUTION, SINE_RESOLUTION)
plt.plot(j,sine_values,'.')
plt.grid()
# Print plain float
np.set_printoptions(suppress=True)

# print(sine)
original_stdout = sys.stdout

with open('sine_values.txt', 'w') as f:
    sys.stdout = f # Change the standard output to the file we created.
    sine_integer = np.round(((2**(SINE_VALUES_BITSIZE-1))-1)*sine).astype(int)
    np.set_printoptions(threshold=sys.maxsize)
    print(np.array2string(sine_integer, separator=',', formatter={'str_kind': lambda x: x}))
    sys.stdout = original_stdout # Reset the standard output to its original value
```
